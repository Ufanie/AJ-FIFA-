```python
import pandas as pd
import numpy as np
df = pd.read_csv('Anie.csv')
df.head()
```

```python
df["Nationality"] = df["Nationality"].astype("category")
df["Positions"] = df["Positions"].astype("category")

#######
df.dropna(inplace=True)
df = df[df["Age"] > 0 ]
###
df.rename(columns={"OVA": "OVA","Hits":"Popularity"},inplace=True)
#####
df.reset_index(drop=True, inplace=True)
###
print(df.head(20))
```

```python
df["Team_Contract"] = df["Team_Contract"].str[6:].replace()
print (df)
```
